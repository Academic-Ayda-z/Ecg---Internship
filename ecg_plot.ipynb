```python

#length most be >3
fake_rpeaks= rpeaks.copy()#[i for i in rpeaks['ECG_R_Peaks']]#np.array([116, 391, 676,720])
#len(arr)-1 == 720
temp=[0,1]+[i for i in rpeaks['ECG_R_Peaks']]
print(temp)
#if len(temp)<3:
  #  temp.append(719)
#if len(temp)<4:
 # temp.append(720)
fake_rpeaks['ECG_R_Peaks']= np.array(temp)
_, waves_peak = nk.ecg_delineate(arr, fake_rpeaks, sampling_rate=721/2, method="peak")
```

```python
plot = nk.events_plot([waves_peak['ECG_T_Peaks'][:], 
                       waves_peak['ECG_P_Peaks'][:],
                       waves_peak['ECG_Q_Peaks'][:],
                       waves_peak['ECG_S_Peaks'][:],
                      fake_rpeaks['ECG_R_Peaks'][:]],
                       arr)

plt.title('T,P,Q,S,R peaks') 
```

```python
# Zooming into the first 3 R-peaks, with focus on T_peaks, P-peaks, Q-peaks and S-peaks
import matplotlib.colors as mcolors
#colors =list(mcolors.BASE_COLORS.values())
colors=[(0.7, 0.7, 0), (0, 0.6, 0), (1, 0, 0), (0, 0.75, 0.75), (0.75, 0, 0.75), (0.75, 0.75, 0), (0, 0, 0), (1, 1, 1)]
labels=['ECG_T_Peaks','ECG_P_Peaks','ECG_Q_Peaks','ECG_S_Peaks','ECG_R_Peaks']
all_waves=[waves_peak['ECG_T_Peaks'][2:], 
                       waves_peak['ECG_P_Peaks'][:],
                       waves_peak['ECG_Q_Peaks'][2:],
                       waves_peak['ECG_S_Peaks'][2:],
                       fake_rpeaks['ECG_R_Peaks'][2:]]
#find ys from all_wacves(xes) and mean while impelement a new xs that do not contain nan(float).
all_xs=[]
all_ys=[]
for row in all_waves:
  tempx=[]
  tempy=[]
  for column in row:
    if type(column)!=type(0.0):
      tempy.append(arr[column])
      tempx.append(column)
  all_ys.append(tempy)
  all_xs.append(tempx)

plt.plot(arr)

for i in range(len(all_waves)):      
  plt.plot(all_xs[i],all_ys[i], ".",color=colors[i],label=labels[i])
  plt.legend(bbox_to_anchor =(0.75, 1.15), ncol = 3)
plt.plot(np.zeros_like(arr), "--", color="black")
plt.show()
print()
```
